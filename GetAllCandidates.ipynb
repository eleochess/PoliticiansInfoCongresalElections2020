# Obteniendo todos los candidatos para las elecciones congresales del 2020

```python
import requests
import csv
import json
```

```python
partidos = []
with open('partidos.csv') as csv_partidos:
    csv_read_partidos = csv.reader(csv_partidos, delimiter=',')
    line_count = 0
    for row in csv_read_partidos:
        if line_count == 0:
            line_count = line_count + 1
        else:
            partidos.append(row)
```

```python
ubigeo = []
with open('ubigeo.csv') as csv_ubigeo:
    csv_read_ubigeo = csv.reader(csv_ubigeo, delimiter=',')
    line_count = 0
    for row in csv_read_ubigeo:
        if line_count == 0:
            line_count = line_count + 1
        else:
            ubigeo.append(row)
```

```python
for ubi in ubigeo:
    for par in partidos:
        print(ubi[0] + " - " + par[1])
```

```python
for ubi in ubigeo:
    for par in partidos:
        print(ubi[1] + " - " + par[2])
```

```python
with open('candidates.csv', 'w', encoding='utf-8', newline='') as csvfile:
    csvfile.truncate()
    first = 1
    for ubi in ubigeo:
        for par in partidos:
            url = "https://votoinformado.jne.gob.pe/voto/Home/CargarCandidatos?TXUBIGEO=" + ubi[1] + "&TXOP=" + par[2]
            print(url)
            response = requests.get(url)
            json_res = response.json()
            filewriter = csv.writer(csvfile, delimiter = ';', quotechar = '|', quoting = csv.QUOTE_MINIMAL)
            if first == 1:
                filewriter.writerow(['ID_CANDIDATO', 'ID_ENCRIPTADO', 'NOMBRES', 'APE_PAT', 'APE_MAT', 'DNI', 'REGION', 'PARTIDO', 'POSICION', 'FOTO'])
                first = 0
            if json_res["result"] == 0:
                continue
            for can in json_res["data"]["Data"]:
                filewriter.writerow([
                    can["IDCANDIDATO"],
                    can["IDENCRIPTADO"],
                    can["TXNOMBRES"],
                    can["TXAPELLIDOPATERNO"],
                    can["TXAPELLIDOMATERNO"],
                    can["TXDOCUMENTOIDENTIDAD"],
                    can["TXPOSTULADISTRITO"],
                    can["TXORGANIZACIONPOLITICA"],
                    can["NUPOSICION"],
                    "https://votoinformado.jne.gob.pe/voto/fotocandidato/" + can["TXDOCUMENTOIDENTIDAD"] + ".JPG"
                ])
```
